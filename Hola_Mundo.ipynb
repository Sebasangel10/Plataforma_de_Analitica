```python
## Importamos las librerias numpy y matplotlib
import numpy as np
from matplotlib import pyplot as plt
```

Marca

```python
#definimos 2 arreglos de dimension 100
ys = 200 + np.random.randn(100)
x= np.arange(0,100, 1)
```

```python
#graficamos los areglos en un plano
plt.plot(x,ys,'-')
plt.fill_between(x,ys,195,where=(ys>195),facecolor='black',alpha=1)

#Agregamos titulo a las graficas
plt.title("Sample Visualization")
plt.xlabel("eje de las x", size = 16,)
plt.ylabel("eje de las y", size = 16,)
plt.show()
```
